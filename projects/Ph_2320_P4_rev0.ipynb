Confident you can improve on it, and recalling that the [Defend Trade Secrets Act of 2016](https://www.govinfo.gov/content/pkg/PLAW-114publ153/pdf/PLAW-114publ153.pdf) states that reverse engineering is NOT an improper mean to acquire trade secrets, and that trade secrets obtained by reverse engineering can become your own trade secrets, you decide to reverse-engineer the treadmill's physics model to calculate Calories. 

The force generated by static friction, $f_s$, is the force that propels a human forward when the human walks. From the definition of work, the work done by the static friction is:

$W_{f_s} = \int_C \vec{f_s} \cdot d\vec{s} = \int_C |\vec{f_s}| |d\vec{s}| \cos{\theta} \;$ , 

where $\theta$ is the angle between the force of friction and the displacement. In our model of friction, the angle between the force of friction and the displacement is zero, and $\cos$ of zero degrees is 1. Therefore,

$W_{f_s} = \int_C |\vec{f_s}| |d\vec{s}| = \int_C |\mu_s\vec{N}| |d\vec{s}| = \mu_s N \int_C ds = \mu_s N \Delta s \;$ , 

where $N$ is the normal force produced by the surface, $\mu_s$ is the coefficient of static friction, and $\Delta s$ is the displacement in the dimension $s$. The *effective* coefficient of static friction assumed by the treadmill is:

$\mu_s = \frac{W_{f_s}}{N \Delta s}$
